```python
#now to determine seasons based on dates and identify seasonal average 
```

```python
df3['race_month']  = df3['race_dates'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
#now classfying months to season
def seas(x):
    if x == 12 or x <= 2:
        return 'Winter'
    elif 3 <= x <= 5:
        return 'Spring'
    elif 6 <= x <= 8:
        return 'Summer'
    else:
        return 'Autumn'

df3['race_season']= df3['race_month'].apply(seas)
```

```python
df3.head(5)
```

```python
#identifying seasonal average
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
